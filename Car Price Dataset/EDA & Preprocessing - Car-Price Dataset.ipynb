## Price
---


* **Here, will check boxplot for detecting outliers and distribution plot to know 
the distribution of price in the car dataset**

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title('Car Price Distribution Plot')
sns.distplot(df['Price'])

plt.subplot(1,2,2)
plt.title('Car Price box Plot')
sns.boxplot(y="Price", data=df)

plt.show()
```

* **From the above chart, we can say that there are outliers in the 'Price' and There is skewness on the right side.**

* **So, we will remove otliers by method of inter-quantitile range**

```python
Q1 = df.Price.quantile(0.25)
Q3 = df.Price.quantile(0.75)
print(Q1, Q3)
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
df = df[(df.Price>lower_limit)&(df.Price<upper_limit)]
df
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title('Car Price Distribution Plot')
sns.distplot(df['Price'])

plt.subplot(1,2,2)
plt.title('Car Price box Plot')
sns.boxplot(y="Price", data=df)

plt.show()
```

**After removing outliers by inter-quantiles range, the new boxplot and distribution plot shown above**

```python
print(f'After Removing outliers in Price ,This DataSet Contains {df.shape[0]} rows & {df.shape[1]} columns.')
```

```python
df = df[df.Mileage != 0]
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title('Mileage Distribution Plot')
sns.distplot(df['Mileage'])

plt.subplot(1,2,2)
plt.title('Mileage box Plot')
sns.boxplot(y="Mileage", data=df)
print(df["Mileage"].shape)
plt.show()
```

* **From the above chart, we can say that there are outliers in the 'Mileage' and There is skewness on the right side.**

```python
Q1 = df.Mileage.quantile(0.25)
Q3 = df.Mileage.quantile(0.75)
print(Q1, Q3)
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
df = df[(df.Levy>lower_limit)&(df.Levy<upper_limit)]
df
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title('Mileage Distribution Plot')
sns.distplot(df['Mileage'])

plt.subplot(1,2,2)
plt.title('Mileage box Plot')
sns.boxplot(y="Mileage", data=df)
print(df["Mileage"].shape)
plt.show()
```
